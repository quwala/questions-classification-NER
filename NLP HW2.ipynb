## Q1.1. Describe the dataset qualitatively

__A quick summary of the paper:__
The task that Li and Roth sought to accomplish is that of classifying some relatively simple (short) questions into 50 categories, representing the "genre" of the answer, or what should be the focus of the answer. Categories are grouped up by 6 "coarse" categories (e.g. entities, humans, locations), and every coarse category contains some number of sub-categories, or "fine" categories (e.g. entities:food, humans:group, location:country).

The dataset they operate on contains pairs of (question, category), with categories being comprised of a coarse category and a fine category within it. The dataset could naturally be defined as multi-label (i.e. there could be a question that appears more than once, each time with a different category assigned to it - we will check this during dataset exploration), however the article states that the taggers were instructed to only assign one label per question. There is a problem of inherent ambiguity between categories in this task, and that will hurt the performance of models on the dataset. 
    
The questions were collected from a number of different sources, and were manually labeled. 
The task of question classification is a sub-task of question answering – once we can classify questions to categories, we can substantially reduce the amount of candidate answers to consider when trying to answer them (the questions).

The outlines of the classifier that was trained for the task are as follows:
The classifier is made out of two classifiers, adjoined in a hierarchal fashion - one for coarse class classification and the other for fine classes. The features to be extracted are POS tags, words (unigrams), chunks (noun phrases, verb phrases etc.), Named Entities, and semantically related words.

```python
#returns 4 lists - the training data,training targets, test data and test targets.
def readDataset():
    f_training=open(TRAINING_DATASET_PATH,'r')
    raw_training=f_training.read()
    f_training.close()

    f_test=open(TEST_DATASET_PATH,'r')
    raw_test=f_test.read()
    f_test.close()

    #targeting a specific error i got on my windows machine.
    if raw_training[:3] == 'ן»¿':
        raw_training=raw_training[3:]
        raw_test=raw_test[3:]

    
    x_train,y_train,x_test,y_test = [],[],[],[]

    #extract training questions and categories (address the case of having multiple colons in a line)
    for sent in raw_training.split('\n'):
        sent_split = sent.split()
        category_string = sent_split[0]
        question_string = ' '.join(sent_split[1:])
        x_train.append(question_string)                           
        y_train.append(category_string)
     
    #extract test questions and categories
    for sent in raw_test.split('\n'):
        sent_split = sent.split()
        category_string = sent_split[0]
        question_string = ' '.join(sent_split[1:])
        x_test.append(question_string)                           
        y_test.append(category_string)
        
        
    return x_train, y_train, x_test, y_test

#for example:
x_train, y_train, x_test, y_test = readDataset() 
print('Question: '+x_train[22])
print('Category: '+y_train[22])
print()
print('Question: '+x_test[22])
print('Category: '+y_test[22])
```

I'll create some data structures to make it easier:

```python
def create_vocabulary_counters():
    vocabulary_per_coarse_label=dict()
    vocabulary_per_fine_label=dict()
    vocabulary_overall = ''

    for category in counter_coarse:
        vocabulary_per_coarse_label[category]=''

    for category in counter_fine:
        vocabulary_per_fine_label[category]=''


    for question,category in zip(x_train,y_train):
        vocabulary_per_fine_label[category]+=question+' '
        vocabulary_per_coarse_label[category.split(':')[0]]+=question+' '
        vocabulary_overall+=question+' '


    for category in counter_coarse:
        counter_vocabulary_per_coarse_label[category] = Counter(nltk.word_tokenize(vocabulary_per_coarse_label[category]))

    for category in counter_fine:
        counter_vocabulary_per_fine_label[category] = Counter(nltk.word_tokenize(vocabulary_per_fine_label[category]))

    counter_vocabulary_overall = Counter(nltk.word_tokenize(vocabulary_overall))
    
    return counter_vocabulary_per_coarse_label , counter_vocabulary_per_fine_label , counter_vocabulary_overall

counter_vocabulary_per_coarse_label , counter_vocabulary_per_fine_label , counter_vocabulary_overall = create_vocabulary_counters()
```

## Q1.3 Baseline classifier

```python
#This method translates spacy's NER labels to our course labels.
def ner_to_coarse_label(ner_label):
    if ner_label in ['PERSON','NORP','ORG']:
        return 'HUM'
    elif ner_label in ['FACILITY','EVENT','PRODUCT','WORK_OF_ART','LAW','LANGUAGE']:
        return 'ENTY'
    elif ner_label in ['GPE','LOC']:
        return 'LOC'
    elif ner_label in ['DATE','TIME','PERCENT','MONEY','QUANTITY','ORDINAL','CARDINAL']:
        return 'NUM'
    else:
        return 'DESC'
```

And the classifier:

```python
#gets a question string, predicts a coarse label
def baseline_classifier(question):
    label = ''
    first = question.split()[0]
    first_and_second = question.split()[0]+' '+question.split()[1]

    if first in ['who','whom','whose','Who','Whom','Whose']:
        return 'HUM'
    elif first == 'where' or first == 'Where':
        return 'LOC'
    
    elif first_and_second == 'How much' or first_and_second == 'How many' or first == 'When':
        return 'NUM'
    
    elif first_and_second == 'What is':
        return 'DESC'
    
    elif 'stand for' in question:
        return 'ABBR'
    
    token = ''
    if 'What' in question:
        token = 'What'
    elif 'what' in question:
        token = 'what'
    elif 'Which' in question:
        token = 'Which'
    elif 'which' in question:
        token = 'which'
    
    if token != '':
        doc = nlp(question)
        ents = [e for e in doc.ents]
        if len(ents) == 0:
            label = random.choice(['LOC','HUM','NUM','ENTY','DESC'])
        elif len(ents) == 1:
            label = ner_to_coarse_label(ents[0].label_)#translates spacy NER to our coarse labels
        else: #more than 1 entity    
            token_index = question.index(token)
            ents_after_token = list(filter(lambda ent: ent.start_char > token_index, ents))

            if len(ents_after_token) == 0:
                label = ner_to_coarse_label(ents[0].label_)
            else:
                label = ner_to_coarse_label(ents_after_token[0].label_)
        
                
    else:
        doc = nlp(question)
        ents = [e for e in doc.ents]
        if len(ents) == 0:
            label = random.choice(['LOC','HUM','NUM','ENTY','DESC']) 
        else:
            label = ner_to_coarse_label(ents[0].label_)#translates spacy NER to our coarse labels


    return label

#for example:
question = 'Where are the Rocky Mountains ?'
print(baseline_classifier(question))

question = 'What river flows between Fargo , North Dakota and Moorhead , Minnesota ?'
print(baseline_classifier(question))

question = 'What position did Willie Davis play in baseball ?'
print(baseline_classifier(question))
```

We note that in question such as: "How far is the service line from the net in Tennis ?", spacy's NER will not tag the word tennis, whereas if it were capitalized (Tennis), it would. correctly (with an EVENT tag).

```python
def plot_confusion_matrix(cm, classes,
                          title,
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
labels=["ABBR","ENTY", "HUM", "NUM", "LOC", "DESC"]

C = confusion_matrix([fine_label.split(':')[0] for fine_label in y_test],
                     [baseline_classifier(question) for question in x_test],
                    labels)
print(C)

plt.figure()
plot_confusion_matrix(C, labels,
                      title='Confusion matrix - Baseline Classifier')
plt.show()
```

```python
#outputs 5 examples of each error (false positive, true negative) for each label type.
def analyzeErrors(questions,predicted_labels,true_labels,labels):
    
    errors = dict(dict()) #errors = dict(label -> dict(error_type -> question string))
    for label in labels:
        errors[label]=dict()
        errors[label]['FP'] = []
        errors[label]['FN'] = []
    
    
    for q,p,t in zip(questions,predicted_labels,true_labels):
        if t!=p:
            if len(errors[p]['FP'])<5:
                errors[p]['FP'].append((q,p,t))
            elif len(errors[t]['FN'])<5:
                errors[t]['FN'].append((q,p,t))

    
    for label in labels:
        print('\n\nFalse positives for label {}:'.format(label))
        for q,p,t in errors[label]['FP']:
            print('Question: {} Predicted label: {} , True label: {}'.format(q,p,t))
            
        print('\n\nFalse negatives for label {}:'.format(label))
        for q,p,t in errors[label]['FN']:
            print('Question: {} Predicted label: {} , True label: {}'.format(q,p,t))
        
    
```

Below is our feature extracting specs, note that this has already been tuned, and produces the best results out of all of our previously trained models. We will show some more tuning of the model in 2.2 (using word embeddings)

```python
BEGINNING_OF_SENTENCE_STRING_IDENTIFIER = '<bos>'
BEGINNING_OF_SENTENCE_INTEGER_IDENTIFIER = -1
END_OF_SENTENCE_STRING_IDENTIFIER = '<eos>'
#We don't extract any integer features, but rather boolean ones (o and 1). So 2 is fine as a unique identifier:
END_OF_SENTENCE_INTEGER_IDENTIFIER = 2 


#gets a sentence (list of strings) and an index into that sentence (word), returns a dictionary of features for sentence[i]
def features(sentence, index):
    word = sentence[index][0]
    pos_tag = sentence[index][1]
    
    features = {
        'word_lowercase':word.lower(),#no need to store the original word form, we have ORT features for that.
        'pos_tag':pos_tag,
        #ORT:
        'is_digit':int(word.isdigit()),#numeric?
        'is_title':int(word.istitle()),#capitalization?
        'is_upper':int(word.isupper()),#all caps?
        'is_punct':int(word in string.punctuation),
        
        'prefix2':word[:2],
        'prefix3':word[:3],
        'suffix3':word[-3:],
        'suffix2':word[-2:],
    }


    #features for previous word in the sentence:
    
    if index > 0:
        prev_word = sentence[index-1][0]
        prev_pos_tag = sentence[index-1][1]
        features.update({
            'prev_word_lowercase': prev_word.lower(),
            'prev_pos_tag' :prev_pos_tag,

            #ORT:
            'prev_is_digit': int(prev_word.isdigit()),
            'prev_is_title': int(prev_word.istitle()),
            'prev_is_upper': int(prev_word.isupper()),
            'prev_is_punct':int(prev_word in string.punctuation),

        })
        
        
    else:#this is the first word in the sentence
        prev_word = BEGINNING_OF_SENTENCE_STRING_IDENTIFIER
        prev_pos_tag = BEGINNING_OF_SENTENCE_STRING_IDENTIFIER
        features.update({
            'prev_word_lowercase': prev_word.lower(),
            'prev_pos_tag' :prev_pos_tag,

            #ORT: assign unique values to indicate beggining of sentence
            'prev_is_digit': BEGINNING_OF_SENTENCE_INTEGER_IDENTIFIER,
            'prev_is_title': BEGINNING_OF_SENTENCE_INTEGER_IDENTIFIER,
            'prev_is_upper': BEGINNING_OF_SENTENCE_INTEGER_IDENTIFIER,
            'prev_is_punct': BEGINNING_OF_SENTENCE_INTEGER_IDENTIFIER,

        })
        
        
    #features for next word in the sentence:

        
    if index < len(sentence)-1:
        next_word = sentence[index+1][0]
        next_pos_tag = sentence[index+1][1]
        
        features.update({
            'next_word_lowercase': next_word.lower(),
            'next_pos_tag' :next_pos_tag,
            
            'next_is_digit':int(next_word.isdigit()),
            'next_is_title':int(next_word.istitle()),
            'next_is_upper':int(next_word.isupper()),
            'next_is_punct':int(next_word in string.punctuation),

        })
        
        
    else:#this is the last word in the sentence.
        next_word = END_OF_SENTENCE_STRING_IDENTIFIER
        next_pos_tag = END_OF_SENTENCE_STRING_IDENTIFIER
        
        features.update({
            'next_word_lowercase': next_word.lower(),
            'next_pos_tag' :next_pos_tag,

            'next_is_digit':END_OF_SENTENCE_INTEGER_IDENTIFIER,
            'next_is_title':END_OF_SENTENCE_INTEGER_IDENTIFIER,
            'next_is_upper':END_OF_SENTENCE_INTEGER_IDENTIFIER,
            'next_is_punct':END_OF_SENTENCE_INTEGER_IDENTIFIER,

        })

                
    return features
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.optim as optim
import torch.nn.functional as F
import nltk
import numpy as np
from collections import Counter
flatten = lambda l: [item for sublist in l for item in sublist]
from sklearn_crfsuite import metrics
random.seed(1024)

USE_CUDA = torch.cuda.is_available()
gpus = [0]
torch.cuda.set_device(gpus[0])

FloatTensor = torch.cuda.FloatTensor if USE_CUDA else torch.FloatTensor
LongTensor = torch.cuda.LongTensor if USE_CUDA else torch.LongTensor
ByteTensor = torch.cuda.ByteTensor if USE_CUDA else torch.ByteTensor

def getBatch(batch_size, train_data):
    random.shuffle(train_data)
    sindex = 0
    eindex = batch_size
    while eindex < len(train_data):
        batch = train_data[sindex: eindex]
        temp = eindex
        eindex = eindex + batch_size
        sindex = temp
        yield batch
    
    if eindex >= len(train_data):
        batch = train_data[sindex:]
        yield batch

def prepare_sequence(seq, word2index):
    idxs = list(map(lambda w: word2index[w] if word2index.get(w) is not None else word2index["<UNK>"], seq))
    return Variable(LongTensor(idxs))

def prepare_word(word, word2index):
    return Variable(LongTensor([word2index[word]]) if word2index.get(word) is not None else LongTensor([word2index["<UNK>"]]))

def prepare_tag(tag,tag2index):
    return Variable(LongTensor([tag2index[tag]]))

corpus = nltk.corpus.conll2002.iob_sents()

data = []
for cor in corpus:
    sent, _, tag = list(zip(*cor))
    data.append([sent, tag])

print(len(data))
print(data[0])


sents,tags = list(zip(*data))
vocab = list(set(flatten(sents)))
tagset = list(set(flatten(tags)))

word2index={'<UNK>' : 0, '<DUMMY>' : 1} # dummy token is for start or end of sentence
for vo in vocab:
    if word2index.get(vo) is None:
        word2index[vo] = len(word2index)
index2word = {v:k for k, v in word2index.items()}

tag2index = {}
for tag in tagset:
    if tag2index.get(tag) is None:
        tag2index[tag] = len(tag2index)
index2tag={v:k for k, v in tag2index.items()}



WINDOW_SIZE = 2
windows = []

for sample in data:
    dummy = ['<DUMMY>'] * WINDOW_SIZE
    window = list(nltk.ngrams(dummy + list(sample[0]) + dummy, WINDOW_SIZE * 2 + 1))
    windows.extend([[list(window[i]), sample[1][i]] for i in range(len(sample[0]))])


    
#a few additions here to make our data consistent with theirs:    
##############################
x_our = [] #our version of the data representation
y_our = []
for sentence in corpus:
    x_our+=get_features(sentence)
    y_our+=get_labels(sentence)
    
bundled = list(zip(windows,x_our,y_our))

random.shuffle(bundled) #randomize the order consistantly through all representations

windows , x_our , y_our = zip(*bundled)

train_data = list(windows[:int(len(windows) * 0.9)])
test_data = list(windows[int(len(windows) * 0.9):])

our_train_x = x_our[:int(len(x_our)*0.9)]
our_train_y = y_our[:int(len(y_our)*0.9)]

our_test_x = x_our[int(len(x_our)*0.9):]
our_test_y = y_our[int(len(y_our)*0.9):]
##############################


class WindowClassifier(nn.Module): 
    def __init__(self, vocab_size, embedding_size, window_size, hidden_size, output_size):

        super(WindowClassifier, self).__init__()
        
        self.embed = nn.Embedding(vocab_size, embedding_size)
        self.h_layer1 = nn.Linear(embedding_size * (window_size * 2 + 1), hidden_size)
        self.h_layer2 = nn.Linear(hidden_size, hidden_size)
        self.o_layer = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()
        self.softmax = nn.LogSoftmax()
        self.dropout = nn.Dropout(0.3)
        
    def forward(self, inputs, is_training=False): 
        embeds = self.embed(inputs) # BxWxD
        concated = embeds.view(-1, embeds.size(1)*embeds.size(2)) # Bx(W*D)
        h0 = self.relu(self.h_layer1(concated))
        if is_training:
            h0 = self.dropout(h0)
        h1 = self.relu(self.h_layer2(h0))
        if is_training:
            h1 = self.dropout(h1)
        out = self.softmax(self.o_layer(h1))
        return out

BATCH_SIZE = 128
EMBEDDING_SIZE = 50 # x (WINDOW_SIZE*2+1) = 250
HIDDEN_SIZE = 300
EPOCH = 3
LEARNING_RATE = 0.001


model = WindowClassifier(len(word2index), EMBEDDING_SIZE, WINDOW_SIZE, HIDDEN_SIZE, len(tag2index))
if USE_CUDA:
    model = model.cuda()
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

for epoch in range(EPOCH):
    losses = []
    for i,batch in enumerate(getBatch(BATCH_SIZE, train_data)):
        x,y=list(zip(*batch))
        inputs = torch.cat([prepare_sequence(sent, word2index).view(1, -1) for sent in x])
        targets = torch.cat([prepare_tag(tag, tag2index) for tag in y])
        model.zero_grad()
        preds = model(inputs, is_training=True)
        loss = loss_function(preds, targets)
        losses.append(loss.data.tolist()[0])
        loss.backward()
        optimizer.step()

        if i % 1000 == 0:
            print("[%d/%d] mean_loss : %0.2f" %(epoch, EPOCH, np.mean(losses)))
            losses = []


for_f1_score = []

accuracy = 0
for test in test_data:
    x, y = test[0], test[1]
    input_ = prepare_sequence(x, word2index).view(1, -1)

    i = model(input_).max(1)[1]
    pred = index2tag[i.data.tolist()[0]]
    for_f1_score.append([pred, y])
    if pred == y:
        accuracy += 1

print(accuracy/len(test_data) * 100)



y_pred, y_test = list(zip(*for_f1_score))
sorted_labels = sorted(
    list(set(y_test) - {'O'}),
    key=lambda name: (name[1:], name[0])
)

y_pred2 = [[y] for y in y_pred] # this is because sklearn_crfsuite.metrics function flatten inputs
y_test2 = [[y] for y in y_test]

print(metrics.flat_classification_report(
    y_test2, y_pred2, labels = sorted_labels, digits=3
))

```
